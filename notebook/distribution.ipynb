## 二项分布

服从二项分布的随机变量X表示在n个独立的是/非试验中成功的次数，其中每次试验的成功概率为p。

$$P(X=k) = \left(\frac{n!}{k!\left(n-k\right)}\right)p^k\left(1-p\right)^{n-k}$$
$$E(X) = np,\ D(X) = np\left(1-p\right)$$

## 正态分布

正态分布是一种连续分布，其函数可以在实线上的任何地方取值。正态分布由两个参数描述：分布的平均值$\mu$和方差$\sigma^2$ 。

正态分布又称高斯分布。若随机变量X服从一个数学期望为$\mu$、方差为$\sigma^2$的正态分布，记为$N(\mu，\sigma^2)$。其概率密度函数为正态分布的期望值$\mu$决定了其位置，其标准差$\sigma$决定了分布的幅度。

$$f(x)=\frac{1}{\sigma\sqrt{2\pi}}exp\left(-\frac{\left(x-\mu\right)^2}{2\sigma^2}\right)$$

当$\mu = 0,\sigma = 1$时的正态分布是标准正态分布。

$$f(x)=\frac{1}{\sqrt{2\pi}}exp\left(-\frac{x^2}{2}\right)$$
$$X \sim N(\mu ,{\sigma ^2})，Y = \frac{X - \mu}{\sigma} \sim N(0, 1)$$

正态分布的一些特性:

- 如果$X \sim N(\mu ,{\sigma ^2})$,且a与b是实数，那么$aX+b \sim N(a \mu + b ,{(a \sigma) ^2})$
- 如果$X \sim N(\mu_X ,{\sigma_X ^2})$与$Y \sim N(\mu_Y ,{\sigma_Y ^2})$是统计独立的正态随机变量,那么
    - $U = X + Y \sim N(\mu_X + \mu_Y, {\sigma_X ^2}+{\sigma_Y ^2})$
    - $V = X - Y \sim N(\mu_X - \mu_Y, {\sigma_X ^2}+{\sigma_Y ^2})$
    - U与V是相互独立的。(要求X与Y的方差相等)
- 如果$X \sim N(0 ,{\sigma_X ^2})$与$Y \sim N(0 ,{\sigma_Y ^2})$是独立常态随机变量，那么它们的积XY服从概率密度函数为p的分布$p(z) = \frac{1}{\pi\sigma_X\sigma_Y}K_0\left(\frac{\left|z\right|}{\sigma_X\sigma_Y}\right)$,其中$K_0$是修正贝塞尔函数（modified Bessel function）
它们的比符合柯西分布，满足$X/Y \sim Cauchy\left(\sigma_X / \sigma_Y\right)$
- 如果$X_1,\cdots, X_n$为独立标准常态随机变量，那么$X_1^2 + \cdots + X_n^2$服从自由度为n的卡方分布。


```python
%matplotlib inline

import math
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def gauss_fun(sigma, mu):
    '''高斯函数，根据给出的标准差sigma与均值mu，给出x,y序列'''
    x = np.linspace(mu - 6 * sigma, mu + 6 * sigma, 250)
    y = np.exp(-(x - mu) ** 2 /(2* sigma **2))/(math.sqrt(2*math.pi)*sigma)
    return x,y

sigma01 = math.sqrt(1)
mu01 = 0
x01, y01 = gauss_fun(sigma01, mu01)

sigma02 = math.sqrt(0.5)
mu02 = 0
x02, y02 = gauss_fun(sigma02, mu02)

sigma03 = math.sqrt(5)
mu03 = 0
x03, y03 = gauss_fun(sigma03, mu03)

sigma04 = math.sqrt(1)
mu04 = -2
x04, y04 = gauss_fun(sigma04, mu04)

p1, = plt.plot(x01, y01, "g-", linewidth=2, label="$\mu=0,\ \sigma^2=1$")
p2, = plt.plot(x02, y02, "b-", linewidth=2, label="$\mu=0,\ \sigma^2=0.5$")
p3, = plt.plot(x03, y03, "m-", linewidth=2, label="$\mu=0,\ \sigma^2=5$")
p4, = plt.plot(x04, y04, "r-", linewidth=2, label="$\mu=-2,\ \sigma^2=1$")

x = np.linspace(-6, 6, 50)
plt.plot(x, stats.norm.pdf(x=x, loc=0, scale=math.sqrt(0.5)), "r*")

plt.legend(handles=[p1,p2,p3,p4], frameon=False)
plt.xlim(xmin=-6, xmax=6)
# plt.xlabel("x")
# plt.ylabel("y")
plt.grid(True)
plt.title("高斯分布")
plt.show()
```

## 泊松分布

一个服从泊松分布的随机变量X，表示在具有比率参数（rate parameter）λ的一段固定时间间隔内，事件发生的次数。参数λ告诉你该事件发生的比率。随机变量X的平均值和方差都是λ。

$$P(X=k) = \frac{\lambda^k}{k!}e^{-\lambda}, \ k = 0,1,\cdots$$
$$E(X) = \lambda,\ D(X) = \lambda$$

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
 
k = range(0,13)
for ld in range(7):
    y = stats.poisson.pmf(k,ld)
    plt.xlabel('K')
    plt.ylabel('P')
    plt.title('泊松分布')
    plt.plot(y,label="$\lambda = %d$" % ld)
 
plt.legend()
plt.show()
```
